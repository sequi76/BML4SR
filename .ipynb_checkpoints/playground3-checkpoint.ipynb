```python
import stan
import pandas as pd
import numpy as np
import scipy.stats as stats 
import matplotlib
import matplotlib.pyplot as plt
import arviz as az
#matplotlib.rcParams['figure.figsize']=5,5
matplotlib.rcParams['figure.dpi']=100
#plt.style.use('ggplot')

# this is a work around found at 
# https://stackoverflow.com/questions/56154176/runtimeerror-asyncio-run-cannot-be-called-from-a-running-event-loop-in-spyd
import nest_asyncio
nest_asyncio.apply()
```

```python
my_model = """
data {
 int <lower=0> N; // number of datapoints
 real mu, sigma;
}

parameters {
  vector [N] x;
  real<lower=0> mu_true; 
  real<lower=0> sigma_true;
}

transformed parameters{
  real mut=0, sigmat=0;
  for (n in 1:N) {
  mut += x[n];
  sigmat += (x[n]-mu)*(x[n]-mu);
  }
  mut = mu*N;
  sigmat = sigma*sigma*N*N;
}

model {
    mu_true ~ normal(0,5);
    sigma_true ~ cauchy(0,5);
    x ~ normal(mu_true, sigma_true);
    }
"""
```

```python
my_data = {'N': 5, 'mu': -1, 'sigma': 8}
```

```python
model = stan.build(my_model, data=my_data, random_seed=0);
```

```python
fit = model.sample(num_chains=4, num_samples=1000, delta=0.8)
```

```python
df = fit.to_frame() 
df
```

```python
plt.hist(df['mu_true'], histtype='step', density=1)
plt.show()
```

```python
plt.scatter(df['mu_true'], df['sigma_true'], s=0.2)
plt.ylim(0,50)
plt.show()
```
