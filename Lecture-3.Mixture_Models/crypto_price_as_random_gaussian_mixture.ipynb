```python
import pandas as pd
import numpy as np
import scipy.stats as stats 
import matplotlib
import matplotlib.pyplot as plt
#matplotlib.rcParams['figure.figsize']=5,5
matplotlib.rcParams['figure.dpi']=100
#plt.style.use('ggplot')
```

## Crypto price as Gaussian Mixture

```python
sigmas = [10,100,1000]
crypto_price = [100000]
#np.random.seed(2)
z = np.random.choice([0,1,2], p=[0.7,0.25,0.01], size=5000)
for zn in z:
    crypto_price.append(crypto_price[-1]+np.random.normal(0,sigmas[zn]))

plt.figure(dpi=100)
plt.plot(crypto_price)
plt.title('Emulated crypto price')
plt.ylabel('USD')
plt.xlabel('Time')
plt.show()
```

```python
for sss in range(20):
    sigmas = [10,100,1000]
    crypto_price = [100000]
    np.random.seed(sss)
    z = np.random.choice([0,1,2], p=[0.7,0.25,0.05], size=5000)
    for zn in z:
        crypto_price.append(crypto_price[-1]+np.random.normal(0,sigmas[zn]))
    
    plt.figure(dpi=100)
    plt.plot(crypto_price)
    plt.title('Emulated crypto price, seed='+str(sss))
    plt.ylabel('USD')
    plt.xlabel('Time')
    plt.show()
```
