```python
import pickle
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
#import stan
import pandas as pd
from scipy import stats
from scipy.special import softmax
from datetime import datetime
import time
import math
import ternary
import mpltern
import random
```

```python
sample0 = np.random.dirichlet([1,1,2], size=300)
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1, projection='ternary')
ax.set_title('Simplex plot')        
ax.scatter([x[0] for x in sample0],[x[1] for x in sample0],[x[2] for x in sample0], s=50, color='magenta', marker='x', label='α=[1, 1, 1]') # play: comment this line if not playing
plt.legend(fontsize=20)
plt.show()
```

```python
sample1 = np.random.dirichlet([70,15,15], size=100)
sample2 = np.random.dirichlet([0.1,0.1,0.1], size=100)
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1, projection='ternary')
ax.set_title('Simplex plot')        
ax.scatter([x[0] for x in sample1],[x[1] for x in sample1],[x[2] for x in sample1], s=50, color='red', marker='x', label='α=[3, 3, 3]') # play: comment this line if not playing
ax.scatter([x[0] for x in sample2],[x[1] for x in sample2],[x[2] for x in sample2], s=50, color='blue', marker='o', label='α=[0.1, 0.1, 0.1]') # play: comment this line if not playing
plt.legend(fontsize=20)
plt.show()
```
